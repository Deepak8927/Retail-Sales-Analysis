## **Step by Step Problem**
**Step 1: Load and Explore the Dataset**

Before analyzing the sales data, we need to load the dataset into Pandas and explore its structure.

‚úÖ Task: Load the CSV File and Display Basic Information

```python
# Solution code for step 1:-
import pandas as pd
data_df = pd.read_csv("Retail_sales_data.csv")
print(data_df.head().to_string(index=False))

data_df.info()
```

**Step 2: Data Cleaning & Preprocessing**

Handle missing values, correct data types, and ensure consistency:

* Replace missing values in numerical columns (Quantity Sold, Unit Price, Total Sales Amount) with mean
* Replace missing values in categorical columns (Payment Method, Customer Segment) with mode
* Convert Date column to datetime
* Check and confirm correct data types using .dtypes

This step ensures the dataset is clean, consistent, and ready for analysis.

```python
# Solution code for step 2 :-

#Replacing missing values in numerical columns with mean
data_df = data_df.fillna(data_df.select_dtypes(include=["number"]).mean().round(2))

#Replacing missing values in categorical columns with mode
string_cols = data_df.select_dtypes(include=['object']).columns   #Selecting all the string values column
data_df[string_cols] = data_df[string_cols].fillna(data_df[string_cols].mode().iloc[0])

#Converting Date column to datetime datatype
data_df['Date'] = pd.to_datetime(data_df['Date'])

data_df.info()
```

**‚úÖ Step 3: Operational Insight**

Identify key business trends that help management optimize operations:
* Most common payment methods used by customers
* Cities with the highest number of orders


üîπ 1. Identify the Most Common Payment Methods used by customers

```python
# Solution code for step 3: Part 1
payment_counts_df = data_df['Payment Method'].value_counts().reset_index()
payment_counts_df.columns = ['Payment Methods', 'No. of times used']
print("The Most Common Payment Methods used by customers:\n")
print(payment_counts_df.to_string(index=False))
```

üîπ 2. Identify Cities with the Highest Order Counts

```python
# Solution code for Step 3: Part 2
cityOrder_counts_df = data_df['City'].value_counts().reset_index()
cityOrder_counts_df.columns = ['Cities', 'Order Counts']
print("Cities with the Highest Order Counts:\n")
print(cityOrder_counts_df.to_string(index=False))
```

**üîπ Step 4 Product Analysis**

Detect trends and inefficiencies in product sales and pricing:

Products with premium pricing but low sales volume

Categories most frequently purchased by different customer segments

üîπ 1. Products with Premium Pricing but Low Sales Volume

üéØ Goal:
Find products that have a high unit price but low total quantity sold ‚Äî possibly overpriced or not in demand.


```python
# Solution code for Step 4 part 1
grouped = data_df.groupby('Product Name').agg({'Unit Price':lambda x: round(x.mean(), 2),'Quantity Sold':'sum'}).reset_index()

sorted_df = grouped.sort_values(by=['Unit Price', 'Quantity Sold'], ascending=[False, True])

print("Top 10 Products with Premium Pricing but Low Sales Volume:\n")
print(sorted_df.head(10).to_string(index=False))
```

üîπ 2. Categories Most Frequently Purchased by Different Customer Segments

üéØ Goal:
See which product categories are most popular for each customer segment (e.g., ‚ÄúCorporate‚Äù, ‚ÄúIndividual‚Äù).




```python
# Solution code for Step 4 part 2
grouped = data_df.groupby(['Customer Segment','Category']).size()
print(grouped)
```

**Step 5: Customer Behavior**

These insights help understand who your customers are, how much they spend, and how often they shop.

Use the dataset to reveal customer behavior patterns:
* Identify top-spending customers by total sales
* Track repeat buyers vs one-time buyers

üîπ 1. Identify Top-Spending Customers by Total Sales

üéØ Goal:
Find customers who have spent the most overall (high-value customers).

```python
# Solution code for Step 5: Part 1
grouped = data_df.groupby('Customer ID')['Total Sales Amount'].sum().reset_index(name='Total Sales Amount')
sorted_grouped = grouped.sort_values(by='Total Sales Amount', ascending= False)
print("Top 10 customers who have spent the most overall (high-value customers):\n")
print(sorted_grouped.head(10).to_string(index=False))
```

üîπ 2. Track Repeat Buyers vs One-Time Buyers

üéØ Goal:
Determine how many customers placed multiple orders vs only one order.

```python
# Solution code for Step 5: Part 2

# one time buyers - customers who have bought for a single time (orderid == 1)
# repeat buyers - customers who have bought for multiple times (orderid > 1).

customer_counts = data_df['Customer ID'].value_counts()

one_time_customers = (customer_counts == 1).sum()
repeat_customers = (customer_counts > 1).sum()

print(f"Number of one-time customers: {one_time_customers}")
print(f"Number of repeat customers: {repeat_customers}")
```

**üîπ Step 6 Data Visualization**

Present your insights using clean and business-relevant visuals:
 1. Growth Line Plot ‚Äì Daily Order Volume Over Time
 2. Vertical Bar Plot ‚Äì Top 10 Products by Total Revenue
 3. Count Plot ‚Äì Orders by Customer Segment

üîπ 1. Growth Line Plot ‚Äì Daily Order Volume Over Time

```python
# Solution code for Step 6: Part 1

dateOrder_counts_df = data_df.groupby('Date').size().reset_index(name='Order Counts')

dates = dateOrder_counts_df['Date'].head(31)
orderCount = dateOrder_counts_df['Order Counts'].head(31)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(dates, orderCount, marker='o')
plt.title("Daily Order Volume Over January Month On Year 2023", fontsize= 16, fontweight = "bold", color = "darkblue", fontstyle = "italic")
plt.xlabel("Each day of January month", fontsize= 12)
plt.xticks(rotation=90)
plt.ylabel("No. of Orders", fontsize= 12)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

üîπ 2. Vertical Bar Plot ‚Äì Top 10 Products by Total Revenue

```python
# Solution code for Step 6: Part 2
grouped_product = data_df.groupby('Product Name')['Total Sales Amount'].sum().reset_index(name='Total Sales Amount')
sorted_grouped_product = grouped_product.sort_values(by='Total Sales Amount', ascending= False)

products = sorted_grouped_product['Product Name'].head(10)
total_sales = sorted_grouped_product['Total Sales Amount'].head(10)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.bar(products ,total_sales, color = "blue", width = 0.5, edgecolor = "black")

plt.title("Top 10 Products by Total Revenue", fontsize= 16, fontweight = "bold", color = "Red", fontstyle = "italic")
plt.xlabel("Products", fontsize = 12)
plt.xticks(rotation=90)
plt.ylabel("Total Sales Amount", fontsize = 12)
plt.grid(axis = "y", linestyle = "--", alpha = 0.5)
plt.show()
```

üîπ 3. Count Plot ‚Äì Orders by Customer Segment



```python
# Solution code for Step 6: Part 3
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 5))
sns.countplot(x='Customer Segment', data=data_df, edgecolor="black")
plt.title('Orders by Customer Segment', fontsize= 16, fontweight = "bold", color = "Green", fontstyle = "italic")
plt.xlabel("Customer Segment", fontsize= 12)
plt.ylabel("No. of Orders", fontsize= 12)
plt.grid(axis = "y", linestyle = "--", alpha = 0.5)
plt.show()
```

üîπ 2. Export Country-wise Average Order Value (AOV)

```python
# Solution code for Step 7: Part 2

grouped_country_df = data_df.groupby('Country')
country_AOV_df = grouped_country_df['Total Sales Amount'].mean().round(2).reset_index(name='Average Order Value')

country_AOV_df.to_csv("Average_order_value.csv", index=False)
```
